```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def draw_losses(path):
    print('Processing .....')
    all_data = []
    files = os.listdir(path)
    for i, file in enumerate(files):
        if i==0:
#             print(file)
            file_path = os.path.join(path, file)
            data = pd.DataFrame(pd.read_csv(file_path, header=None))
            data.columns = ['epoch','total_iters','nLabels','loss']
        elif i!=0:
#             print(file)
            file_path = os.path.join(path, file)
            df2 = pd.DataFrame(pd.read_csv(file_path, header=None))
            df2.columns = ['epoch','total_iters','nLabels','loss']
            data = pd.concat([data,df2], axis=1)
            
# -----------------------------------------------------        
    losses = data[['loss']]
    losses.shape
    losses.fillna(" ")
    
    print('Assigining image names as column names.....')
    column_names = []
    for i in range(0, losses.shape[1]):
        column_names.append('loss_'+str(i))
#     print(column_names)

    losses.columns=column_names

    cols = losses.columns
#     print(cols)
    print('graphs are in progress.....')
    
    losses.cumsum()
    plt.figure()
    # plt.legend().remove
    losses.plot(legend=0, figsize=(8,6),xlabel='Epochs', ylabel='Image wise loss', grid=1)
    plt.show()
    losses_mean = losses.mean(axis=1)
    losses_mean.cumsum()
    losses_mean.plot(legend=0, figsize=(8,6),xlabel='Epochs', ylabel='Average loss', grid=1)
    plt.show()
    
    print('Task completed.....')
#     return losses_mean                               #, losses

            
#     break
```

```python
path_CT = 'D:\AI\code\Azure_notebook_data\medical images predictions\outputs\CT_BH'
CT_mean = draw_losses(path=path_CT)
print('Mean of CT loss')
```

```python
path_DCT = 'D:\AI\code\Azure_notebook_data\medical images predictions\outputs\DynaCT_BH'
draw_losses(path=path_DCT)
```

# Averge loss of last epochs

```python
def frequency_graph(path):
    data_ct = pd.DataFrame(pd.read_csv(path, header=None))
    data_ct.columns = ['image_name','max_epochs', 'loss']
    print(data_ct.describe())
    
    ct_max_epochs = data_ct['max_epochs'].value_counts()
    fig = plt.figure(figsize=(16,6))
    ax = fig.gca()
    ct_max_epochs.plot(kind='bar')
    ax.set_title('Frequency of epochs untill 3 class convergance')
    ax.set_xlabel('Number of epochs')
    ax.set_ylabel('Frequency')
    plt.show()
    
```
